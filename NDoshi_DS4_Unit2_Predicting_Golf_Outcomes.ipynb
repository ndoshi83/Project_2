```python
# Load required directories
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#show lots of columns and rows in pandas
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 3000)
pd.set_option('display.width', 3000)
```

```python
# Load data
df2019 = pd.read_csv('2019_data.csv')
df_historical = pd.read_csv('PGA_Historical_Wide.csv')

df2019.shape, df_historical.shape
```

```python
# Transpose dataframe
df2019_tp = df2019.set_index(['Player Name', 'Variable', 'Date'])['Value'].unstack('Variable').reset_index()

df2019_tp.shape
```

```python
# Creating test set to check accuracy of the model
df2019 = df2019_tp.loc[df2019_tp['Date']== 'June 13 - June 16 (2019)']

# Reset index
df2019 = df2019.reset_index(drop=True)

# Change player name column to match
df2019 = df2019.rename(columns={'Player Name':'Player_Name'})

df2019.shape
```

```python
# Create wrangle function
def wrangle(X):
    # Create copy so original data is not lost
    X_Clean = X.copy()
    # Use relevant columns for analysis
    k_columns = ['1-Putts per Round - (AVG)','2-Putts per Round - (AVG)','3+ Putts per Round - (AVG)',
                 '3-Putt Avoidance - (%)','Accuracy Rating - (RATING)','Ball Speed - (AVG.)',
                 'Carry Distance - (AVG.)','Club Head Speed - (AVG.)','Driving Distance - (AVG.)',
                 'Greens in Regulation Percentage - (%)','Greens or Fringe in Regulation - (%)','Hit Fairway Percentage - (%)',
                 'Left Tendency - (%)','Putting Average - (AVG)','Right Tendency - (%)',
                 'Rough Tendency - (%)','Sand Save Percentage - (%)','Scoring Average - (AVG)',
                 'Scrambling - (%)','Stroke Differential Field Average - (STROKE DIFFERENTIAL)','Victory Leaders - (VICTORIES)']
    X_Clean = X_Clean[k_columns]
    
    # Rename columns to simplify dfs
    X_Clean = X_Clean.rename(columns ={'1-Putts per Round - (AVG)': '1_putts','2-Putts per Round - (AVG)':'2_putts',
                                       '3+ Putts per Round - (AVG)':'3_putts','3-Putt Avoidance - (%)':'3_putt_avoidance',
                                       'Accuracy Rating - (RATING)':'accuracy_rating','Ball Speed - (AVG.)':'ball_speed',
                                       'Carry Distance - (AVG.)':'carry','Club Head Speed - (AVG.)':'club_head_speed',
                                       'Driving Distance - (AVG.)':'driving_dist','Greens in Regulation Percentage - (%)':'GIR',
                                       'Greens or Fringe in Regulation - (%)':'GFIR','Hit Fairway Percentage - (%)':'FIR',
                                       'Left Tendency - (%)':'left','Putting Average - (AVG)':'putting_avg',
                                       'Right Tendency - (%)':'right','Rough Tendency - (%)':'rough','Sand Save Percentage - (%)':'sand_saves',
                                       'Scoring Average - (AVG)':'scoring_avg','Scrambling - (%)':'scrambling',
                                       'Stroke Differential Field Average - (STROKE DIFFERENTIAL)':'stroke_dif','Victory Leaders - (VICTORIES)':'wins'})
    
    # Fill in Nan's with 0 to remove blanks
    X_Clean = X_Clean.fillna(0)
    
    # Convert E to 0
    X_Clean = X_Clean.replace('E',0)
    
    # Reset index
    X_Clean = X_Clean.reset_index(drop=True)
    
    # Convert all except name feature to float
    X_Clean = X_Clean.astype('float')
    
    # Return wrangle dataset
    return X_Clean

train = wrangle(df_historical)
test = wrangle(df2019)

train.shape, test.shape
```

```python
# Create X and y matrices of train and test
target = 'wins'
X_train = train.drop(columns=target)
y_train = train[target]
X_test = test.drop(columns=target)
y_test = test[target]

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# Import libraries needed for model building
# Fit into XGBoost model
import category_encoders as ce
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
```

```python
# Create model
processor = make_pipeline(
    ce.OrdinalEncoder(), 
    SimpleImputer(strategy='mean'),
    LogisticRegression(solver='lbfgs', multi_class='auto', max_iter=10000)
)

processor.fit(X_train, y_train)
```

```python
# Predict the model on the test set
y_pred = processor.predict(X_test)
print('The accuracy score of logistic regression is:\n',processor.score(X_test,y_test))
```

```python
# Pull coefficients for logistic regression
processor.named_steps

# Pull information from pipeline
encoder = processor.named_steps['ordinalencoder']
lr = processor.named_steps['logisticregression']
feat_names = encoder.transform(X_test).columns
lr_coef = pd.Series(lr.coef_[0], feat_names)

# Plot coefficients
n = 25
plt.figure(figsize=(15, 20))
plt.title(f'Top coefficients')
lr_coef.sort_values()[-n:].plot.barh();
plt.savefig('Coefs.png')
```
